```python
#Code:
#Importing Libraries

from sklearn.datasets import load_iris
import numpy as np
from matplotlib import pyplot as plt

#Importing Data
Data = load_iris()
X = Data.data
y = Data.target
X = X[:100, :]
y = y[:100]
print(X.shape)
print(y.shape)

#Train Test Split

indices = np.random.permutation(100)
print(indices)
X1 = X[indices]
y1 = y[indices]
print(y1)
X_train = X1[:70,:]
X_test = X1[70:, :]
y_train = y1[:70]
y_test = y1[70:]
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

#KNN Implementation

predicted = []
for j in range(X_test.shape[0]):
  distances = []
  x2 = X_test[j,:]
  for i in range(X_train.shape[0]):
    x1 = X_train[i,:]
    d=np.sqrt(np.sum((x2-x1)**2))
    distances.append(d)
  #print(distances)
  indices=np.argsort(distances)
  #print(indices)
  selected_ind = indices[:5]
  selected_label = y_train[selected_ind]
  s=np.sum(selected_label)
  if(s >= 3):
    unknown_label = 1
  else:
    unknown_label = 0
  predicted.append(unknown_label)
acc = np.sum(predicted==y_test)/(y_test.shape[0])
print(acc)
```

```python
#Code:
from sklearn.datasets import load_boston
import numpy as np
from matplotlib import pyplot as plt

#Boston Housing Price dataset
Data = load_boston()
X = Data.data
y = Data.target
print(X.shape)
print(y.shape)

#Manually splitting into train and test
X_train = X[:354,:]
X_test = X[354:, :]
y_train = y[:354]
y_test = y[354:]
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

#Predicting the price of samples in the test set
predicted = []
for j in range(X_test.shape[0]):
  distances = []
  x2 = X_test[j,:]
  for i in range(X_train.shape[0]):
    x1 = X_train[i,:]
    d=np.sqrt(np.sum((x2-x1)**2))
    distances.append(d)
  #print(distances)
  indices=np.argsort(distances)
  #print(indices)
  selected_ind = indices[:5]
  selected_label = y_train[selected_ind]
  pred = np.mean(selected_label)
  predicted.append(pred)
```
